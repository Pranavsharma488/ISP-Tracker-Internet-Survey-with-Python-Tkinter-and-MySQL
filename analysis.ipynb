```python
import pandas as pd

file_path = 'Internet Speed vs. Customer Satisfaction Survey .csv'
data = pd.read_csv(file_path)
data.head()
```

```python
data.info()
```

```python
data=data.fillna(method='ffill')
```

```python
data.columns = data.columns.str.strip()


user_count = data['Internet Service Provider'].value_counts()


highest_user_provider = user_count.idxmax()
highest_user_count = user_count.max()

average_satisfaction = data.groupby('Internet Service Provider')['Overall, how satisfied are you with your internet service provider?'].mean()

print("Number of users by Internet Service Provider:\n", user_count)
print(f"\nProvider with the highest users: {highest_user_provider} ({highest_user_count} users)")
print("\nAverage satisfaction by Internet Service Provider:\n", average_satisfaction)
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
user_count.plot(kind='barh', color='skyblue')
plt.title('Number of Users by Internet Service Provider')
plt.ylabel('Internet Service Provider')
plt.xlabel('Number of Users')

plt.subplot(1, 2, 2)
average_satisfaction.plot(kind='bar', color='salmon',width=0.2)
plt.title('Average Satisfaction by Internet Service Provider')
plt.xlabel('Internet Service Provider')
plt.ylabel('Average Satisfaction Rating')


plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np


age_counts = data['Age'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
age_counts.plot(kind='barh', color='skyblue', edgecolor='black', width=0.6)
plt.title('Distribution of Age Groups', fontsize=16)
plt.xlabel('Number of Respondents', fontsize=12)
plt.ylabel('Age Group', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
